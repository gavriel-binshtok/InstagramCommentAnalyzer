```python
import nltk
nltk.download('vader_lexicon')
```

```python
new_words = {
    'queen': 9.0,
    'fire': 1.0,
    'purple_heart': 4.0,
    'smiling': 1.0,
    'pleading': 1.0,
}

sid = SentimentIntensityAnalyzer()

sid.lexicon.update(new_words)
```

```python
list = []
for index, row in data.iterrows():
    ss = sid.polarity_scores(row['comment'])
    list.append(ss)
se = pd.Series(list)
data['polarity'] = se.values

#Display the Sentiment Results
display(data[['comment','polarity']][0:31])
```
